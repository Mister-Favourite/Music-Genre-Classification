```python
#Add whatever you need to import up here.  Or anywhere really

import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("train.csv")
dataset.head(5)
```

```python
#The instrumentalness column has NaN values, which might mess with values
#I'm setting them to 0 for now, but the max value is 0.996 in that column so setting
#it higher than that might make sense in the future.

#Apply to other columns if needed
def set_NaN(x):
    if math.isnan(x):
        return 0
    else:
        return x
```

```python
dataset["instrumentalness"] = dataset["instrumentalness"].apply(lambda x: set_NaN(x))
dataset["key"] = dataset["key"].apply(lambda x: set_NaN(x))

#Use this to filter features that you might not want
feature_columns = ['Popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
                    'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_in min/ms', 'time_signature']

#Feel free to adjust the test_size, random_state, and columns that should be used.
X_train, X_test, y_train, y_test = train_test_split(dataset[feature_columns], dataset['Class'], 
                                                    test_size=0.2, random_state=42)

#If the model that you are using doesn't split the training set into a validation set, here is a basic holdout set
#for validation that is commented out by default
#X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```
